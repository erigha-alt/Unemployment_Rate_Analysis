```python
unemploynment_2020['Month'] = unemploynment_2020['Date'].dt.month
print(unemploynment_2020)
```

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=unemploynment_2020)
plt.show()
```

the estimated employment rate is high in  the south

```python
before_lock = unemploynment_2020[(unemploynment_2020['Month'] >= 1) & (unemploynment_2020['Month'] <= 3)][['Region', 'Estimated Unemployment Rate']]
after_lock = unemploynment_2020[(unemploynment_2020['Month'] >= 3) & (unemploynment_2020['Month'] < 6)][['Region', 'Estimated Unemployment Rate']]

# Compute the mean unemployment rate before lockdown for each region
before_lock = before_lock.groupby('Region')['Estimated Unemployment Rate'].mean().reset_index().rename(
    columns={'Estimated Unemployment Rate':'Unemployment Rate before Lock-Down'})

# Compute the mean unemployment rate after lockdown for each region
after_lock = after_lock.groupby('Region')['Estimated Unemployment Rate'].mean().reset_index().rename(
    columns={'Estimated Unemployment Rate':'Unemployment Rate after Lock-Down'}) 

# Calculate the percentage change in unemployment rate
before_lock['Percentage Change in Unemployment'] = round((after_lock['Unemployment Rate after Lock-Down'] - before_lock['Unemployment Rate before Lock-Down']) / before_lock['Unemployment Rate before Lock-Down'], 2)

# Sort the data by percentage change in unemployment rate
plot_unemploynment_2020 = before_lock.sort_values('Percentage Change in Unemployment', ascending=False)

# Plotting
plt.figure(figsize=(12, 10))
sns.barplot(data=plot_unemploynment_2020, y='Region', x='Percentage Change in Unemployment')
plt.xlabel('Percentage Change in Unemployment')
plt.ylabel('Region')
plt.title('Percentage Change in Unemployment Rate by Region')
plt.show()
```

```python
# Filter data for months 1 to 3 (before lockdown)
before_lock = unemploynment_2020[(unemploynment_2020['Month'] >= 1) & (unemploynment_2020['Month'] <= 3)][['States', 'Estimated Unemployment Rate']]

# Filter data for months 3 to 5 (after lockdown)
after_lock = unemploynment_2020[(unemploynment_2020['Month'] >= 3) & (unemploynment_2020['Month'] < 6)][['States', 'Estimated Unemployment Rate']]

# Compute the mean unemployment rate before lockdown for each region
before_lock = before_lock.groupby('States')['Estimated Unemployment Rate'].mean().reset_index().rename(
    columns={'Estimated Unemployment Rate':'Unemployment Rate before Lock-Down'})

# Compute the mean unemployment rate after lockdown for each region
after_lock = after_lock.groupby('States')['Estimated Unemployment Rate'].mean().reset_index().rename(
    columns={'Estimated Unemployment Rate':'Unemployment Rate after Lock-Down'}) 

# Calculate the percentage change in unemployment rate
before_lock['Percentage Change in Unemployment'] = round((after_lock['Unemployment Rate after Lock-Down'] - before_lock['Unemployment Rate before Lock-Down']) / before_lock['Unemployment Rate before Lock-Down'], 2)

# Sort the data by percentage change in unemployment rate
plot_unemploynment_2020 = before_lock.sort_values('Percentage Change in Unemployment', ascending=False)

# Plotting
plt.figure(figsize=(12, 10))
sns.barplot(data=plot_unemploynment_2020, y='States', x='Percentage Change in Unemployment')
plt.xlabel('Percentage Change in Unemployment')
plt.ylabel('States')
plt.title('Percentage Change in Unemployment Rate by States')
plt.show()
```
